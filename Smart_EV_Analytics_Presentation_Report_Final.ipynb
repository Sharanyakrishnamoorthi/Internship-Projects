# 4. Model Creation

## Clustering Algorithms

**Why Clustering?**

• Groups similar locations together

• Ensures balanced coverage across Chennai

• Considers multiple factors simultaneously

• Handles large datasets efficiently

**Three Algorithms Tested:**

### K-Means Clustering
• Creates exactly 12 balanced groups

• Selects best location from each group

• Fast and reliable results

### DBSCAN Clustering  
• Finds natural density-based clusters

• Handles outliers automatically

• Variable cluster sizes

### Hierarchical Clustering
• Creates tree-like grouping structure

• Shows relationships between areas

• Flexible cluster numbers

**Pre-processing:**
• StandardScaler: Made all features equal importance

• 5 features used: demand, population, coordinates, cost
